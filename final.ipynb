```python
import pandas as pd
df = pd.read_csv('tb.csv')
df.head()
```

```python
df = pd.melt(df, id_vars=['country', 'year'])
df
```

```python
df['gender'] = df.variable.str[0] 
df['age'] = df.variable.str[1:]
df.head()
```

```python
df = df.drop(['variable'], axis = 1)
df.head()
```

```python
df.rename(columns={'value': 'cases'}, inplace=True)
df
```

```python
df = df[df.cases.notnull()]
df
```

```python
df = df.fillna(0)
df.head()
```

```python
df = df.replace({'04' : '0-4', '514' : '5-14', '1524' : '15-24', '2534' : '25-34', '3544' : '35-44', '4554' : '45-54', '5564' : '55-64', '65' : '65+', '014' : 'nan', 'u' : 'nan'})
```
